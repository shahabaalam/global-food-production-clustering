# Exploratory Data Analysis (EDA)

```python
#Plot for annual produce of different countries with quantity in y-axis and years in x-axis
area_list = list(df['Area'].unique())
year_list = list(df.iloc[:,10:].columns)

plt.figure(figsize=(24,12))
for ar in area_list:
    yearly_produce = []
    for yr in year_list:
        yearly_produce.append(df[yr][df['Area'] == ar].sum())
    plt.plot(yearly_produce, label=ar)
plt.xticks(np.arange(53), tuple(year_list), rotation=60)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=8, mode="expand", borderaxespad=0.)
plt.savefig('p.png')
plt.show()
```

# Feature Engineering

```python
mean_produce = []
for i in range(174):
    mean_produce.append(new_df.iloc[i,:].values.mean())
new_df['Mean_Produce'] = mean_produce

new_df['Rank'] = new_df['Mean_Produce'].rank(ascending=False)

new_df.head()
```

```python
#create another dataframe with items and their total production each year from 1961 to 2013
item_list = list(df['Item'].unique())

item_df = pd.DataFrame()
item_df['Item_Name'] = item_list

for yr in year_list:
    item_produce = []
    for it in item_list:
        item_produce.append(df[yr][df['Item']==it].sum())
    item_df[yr] = item_produce
```

```python
item_df.head()
```

```python
sum_col = []
for i in range(115):
    sum_col.append(item_df.iloc[i,1:].values.sum())
item_df['Sum'] = sum_col
item_df['Production_Rank'] = item_df['Sum'].rank(ascending=False)

item_df.head()
```

# K-means clustering

```python
new_df.head()
```

```python
import pandas as pd

# Assuming 'new_df' is your DataFrame and 'year_list' contains column names
X = new_df.iloc[:, :-2].copy()  # Selecting all columns except the last two
X = pd.DataFrame(X)  # Convert to DataFrame if not already

# Convert columns to numeric type
X = X.apply(pd.to_numeric, errors='coerce')

# Assign column names from 'year_list'
X.columns = year_list
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Assuming X is your DataFrame containing numeric data
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X.values)  # Use .values to get the underlying NumPy array

# If you need to convert X back to a NumPy array
X_array = X.values
```

```python
import matplotlib.pyplot as plt

# Assuming X is your DataFrame containing numeric data
plt.scatter(X.iloc[y_kmeans == 0, 0], X.iloc[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1 (Others)')
plt.scatter(X.iloc[y_kmeans == 1, 0], X.iloc[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2 (China, USA, India)')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of countries by Productivity')
plt.xlabel('Feature 1')  # Update with appropriate feature names if needed
plt.ylabel('Feature 2')  # Update with appropriate feature names if needed
plt.legend()
plt.show()
```

# Hierarchical Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram

# Assuming X is your DataFrame containing numeric data
# Example of generating sample data
np.random.seed(42)
X = pd.DataFrame(np.random.rand(20, 4), columns=['Feature1', 'Feature2', 'Feature3', 'Feature4'])

# Perform hierarchical clustering
linkage_matrix = linkage(X, method='ward', metric='euclidean')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, orientation='top', labels=X.index, distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Import necessary libraries
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import numpy as np

# Assuming X_array is your data matrix; ensure it's defined properly
# X_array should be a 2D array-like structure with your features
# Example: X_array = np.array([[1, 2], [2, 3], [3, 4], ...])

# Perform Agglomerative Hierarchical Clustering
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X_array)

# Evaluate clustering
silhouette_avg_hc = silhouette_score(X_array, y_hc)
db_index_hc = davies_bouldin_score(X_array, y_hc)

print(f'Silhouette Score: {silhouette_avg_hc}')
print(f'Davies-Bouldin Index: {db_index_hc}')

# Visualizing clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_array[:, 0], X_array[:, 1], c=y_hc, cmap='viridis')
plt.title('Clusters of Countries by Productivity (Hierarchical Clustering)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
```
